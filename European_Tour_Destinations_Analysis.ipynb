# **Rename The Column**

```python
df = df.rename(columns={'Famous Foods': 'Famous cousine'})
df.head()
```

### **Excluding "spring" from the column. # [data slicing] **bold text**

```python
df['Best session to Visit'] = df['Best Time to Visit'].str[0:6]
df.head()
```

# **Excluding "Spring" from the column [best time to visit]**

```python
df['Best Time to Visit'] = df['Best Time to Visit'].str.replace('Spring', '', regex=False)
df.head()
```

## **Renaming the Column from 'Best time to Visit': 'Best Months to Visit' **bold text**

```python
df = df.rename(columns={'Best time to Visit': 'Best Months to Visit'})
df.head()
```

# **Reviewing the data**

```python
df.head()
```

# **Spliting the Column**

```python
df = ps.split_column(df,'Best Months to Visit',delimiter =" ~ ")
df.head()
```

# **Renaming the column**

```python
df = df.rename(columns={'Best Months to Visit_1': 'Best Months to Visit'})
df.head()
```

# **Replace String From Column**

```python
df['Best Months to Visit'] = df['Best Months to Visit'].str.replace(r'[or Fall]', '', regex=True)

df.head()
```

# **Add String in column**

```python
df['Best session to Visit'] = df['Best session to Visit'].apply(
    lambda x: '- Fall ' + x if 'Fall' not in x else x)
df.head()
```

# **Seprate the session in different Columnn**

```python
df['Best session to Visit'] = df['Best session to Visit'].str.replace('- Fall Spring', 'Spring - Fall')
df.head()
```

## **Display the correlation between Latitude and Longitude for destinations to see their density: Bivariate Graph Analysis**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Longitude', y='Latitude', alpha=0.6)

# Customize the plot
plt.title('Scatter Plot of Destinations by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
```

# **Proportion of Destination By Region: Univariate Graph Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of destinations per region
region_counts = df['Region'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)

# Customize the plot
plt.title('Proportion of Destinations by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# **Cost Of Living Box Plot Interactive Univariate Graph Analysis**

```python
import pandas as pd
import plotly.express as px

# Sample data creation (replace this with your actual DataFrame)
data = {
    'Destination': ['Rome', 'Florence', 'Venice', 'Milan', 'Naples',
                    'Cinque Terre', 'Turin', 'Amalfi Coast', 'Pisa', 'Lake Como'],
    'Cost of Living': ['High', 'Medium', 'Low', 'Medium', 'High',
                       'Low', 'Medium', 'Low', 'High', 'Medium']
}

df = pd.DataFrame(data)

# Convert categorical data to a numerical format for box plot
df['Cost of Living Numeric'] = df['Cost of Living'].map({'Low': 1, 'Medium': 2, 'High': 3})

# Create the interactive box plot with destination labels
fig = px.box(df, y='Cost of Living Numeric',
             x='Destination',  # Set Destination as x-axis
             title='Box Plot of Cost of Living by Destination',
             labels={'Cost of Living Numeric': 'Cost of Living', 'Destination': 'Destination'},
             points="all")  # Add points to show individual data points

# Show the figure
fig.show()
```
